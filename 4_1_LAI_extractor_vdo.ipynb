```python
import os
import cv2
```

```python
DIRECTORY = './images/videos/'
file_list = os.listdir(DIRECTORY)
dataset_list = [file for file in file_list if file.startswith('LAI2_OVER')]
dataset_list.sort()
```

```python
for FILENAME in dataset_list:
    vidcap = cv2.VideoCapture('./images/videos/' + FILENAME)
    count = 0
    os.mkdir('./images/%s' % str(FILENAME.split('.')[0]))
    
    success = True
    while success:
        success, image = vidcap.read()
        cv2.imwrite('./images/%s/frame_%04d.jpg' % (FILENAME.split('.')[0], count), image)

        count += 1

    vidcap.release
```
